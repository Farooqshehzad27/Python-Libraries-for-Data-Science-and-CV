# Reading and Displaying Images in OpenCV and Matplotlib

In this notebook, we will read and display images using two methods:  
1. **OpenCV (`cv2.imshow()`)** - Opens images in a separate window.  
2. **Matplotlib (`plt.imshow()`)** - Displays images inside the notebook.  

ðŸ“Œ **Key Differences:**  
- OpenCV loads images in **BGR format**, while Matplotlib expects **RGB format**.  
- OpenCV requires `cv2.waitKey(0)` to keep the window open until a key is pressed.  
- Matplotlib does not open a separate window; instead, it embeds the image directly into the notebook.  


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the image
img = cv2.imread("images/rose.tif")

```

```python
# Display image using OpenCV
cv2.imshow("Image Display - OpenCV", img)
cv2.waitKey(0)  # Wait for a key press
cv2.destroyAllWindows()  # Close the window
```

```python
#Display Image using Matplotlib

# Convert BGR to RGB for Matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display using Matplotlib
plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.axis("off")  # Hide axis
plt.title("Image Display - Matplotlib")
plt.show()
```

# Basic Image Processing in OpenCV

In this notebook, we will cover basic image processing techniques using OpenCV, including:
- Converting images to grayscale
- Resizing images
- Applying Gaussian Blur
- Edge Detection using Laplacian

These operations are commonly used in computer vision applications for preprocessing images before analysis.


```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread("images/peppers.jpg")

# Check if the image is loaded correctly
if img is None:
    print("Error: Image not found or cannot be loaded.")
```

```python
# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Convert original image to RGB for Matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display images side by side using Matplotlib
plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Original Image")

# Grayscale Image
plt.subplot(1, 2, 2)
plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.title("Grayscale Image")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Resize the image to 400x400 pixels
resized = cv2.resize(img, (400, 400))

# Convert BGR to RGB for proper Matplotlib display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
resized_rgb = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)

# Display images side by side using Matplotlib
plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Original Image")

# Resized Image
plt.subplot(1, 2, 2)
plt.imshow(resized_rgb)
plt.axis("off")
plt.title("Resized Image (400x400)")

plt.show()
```

```python
# Apply Gaussian Blur
blurred = cv2.GaussianBlur(img, (21, 21), 0)

# Convert original and blurred images to RGB for Matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
blurred_rgb = cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB)

# Display images side by side using Matplotlib
plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Original Image")

# Gaussian Blurred Image
plt.subplot(1, 2, 2)
plt.imshow(blurred_rgb)
plt.axis("off")
plt.title("Gaussian Blurred Image")

plt.show()
```

# Edge Detection in OpenCV

Edge detection is an important technique in computer vision that helps in identifying object boundaries within an image.  
In this notebook, we will explore different edge detection techniques using OpenCV, including:

- **Laplacian Edge Detection**
- **Sobel Edge Detection**
- **Canny Edge Detection**

We will visualize and compare these methods using Matplotlib.

```python
# Apply Laplacian edge detection
laplacian = cv2.Laplacian(gray, cv2.CV_64F)

# Display the result
plt.figure(figsize=(6,6))
plt.imshow(laplacian, cmap="gray")
plt.axis("off")
plt.title("Laplacian Edge Detection")
plt.show()
```

```python
# Apply Sobel edge detection
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)  # X-direction
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)  # Y-direction

# Display Sobel X and Sobel Y
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(sobelx, cmap="gray")
plt.axis("off")
plt.title("Sobel X Edge Detection")

plt.subplot(1,2,2)
plt.imshow(sobely, cmap="gray")
plt.axis("off")
plt.title("Sobel Y Edge Detection")

plt.show()
```

```python
# Apply Canny edge detection
canny = cv2.Canny(gray, 50, 150)

# Display the result
plt.figure(figsize=(6,6))
plt.imshow(canny, cmap="gray")
plt.axis("off")
plt.title("Canny Edge Detection")
plt.show()
```

## Summary of Edge Detection Methods

- **Laplacian Edge Detection:** Uses the second derivative to detect rapid intensity changes.
- **Sobel Edge Detection:** Computes the gradient of image intensity separately in X and Y directions.
- **Canny Edge Detection:** Applies Gaussian blur, calculates gradients, and detects strong edges using thresholds.

Each method has its own advantages and is used in different applications in computer vision.


# Image Thresholding in OpenCV

Thresholding is a technique used in image processing to segment images by converting grayscale images into binary images.  
It helps in extracting important features by setting pixel intensity values to either 0 or 255 based on a threshold value.

In this notebook, we will explore thresholding using OpenCV:

- **Simple Thresholding**

We will visualize and compare these methods using Matplotlib.


```python
# Read the image
img = cv2.imread("images/beans.JPG", cv2.IMREAD_GRAYSCALE)  # Load in grayscale

# Display the original grayscale image
plt.figure(figsize=(6,6))
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.title("Original Grayscale Image")
plt.show()
```

```python
# Apply simple thresholding (threshold value = 127)
threshold_value, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
threshold_value, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)

# Display the results
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(thresh1, cmap="gray")
plt.axis("off")
plt.title("Binary Thresholding")

plt.subplot(1,2,2)
plt.imshow(thresh2, cmap="gray")
plt.axis("off")
plt.title("Binary Inverted Thresholding")

plt.show()
```
