```python
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Create line plot with multiple lines
plt.plot(x, y1, label='Prime Numbers', color='r', linestyle='--', marker='o')   # Dashed red line with circle markers
plt.plot(x, y2, label='Other Numbers', color='#33A1C9', linestyle='-.', marker='x')  # Dash-dot blue line with x markers

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Plot Example')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Save the plot to a file
plt.savefig('simple_line_plot.png')

# Show the plot
plt.show()
```

```python
# Figure and Axes

# Sample data
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]
y2 = [i**1.5 for i in x]

# Create figure and subplots
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(10, 6))

# Top Left subplot
axes[0, 0].plot(x, y, color='red')
axes[0, 0].set_title('Top Left')
axes[0, 0].set_ylabel('Y-axis')
axes[0, 0].grid(True)

# Top Right subplot
axes[0, 1].plot(x, y2, color='blue')
axes[0, 1].set_title('Top Right')
axes[0, 1].grid(True)

# Bottom Left subplot
axes[1, 0].plot(x, y[::-1], color='green')
axes[1, 0].set_title('Bottom Left')
axes[1, 0].set_xlabel('X-axis')
axes[1, 0].set_ylabel('Y-axis')
axes[1, 0].grid(True)
axes[1, 0].set_aspect('equal')

# Bottom Right subplot - text and no axes
axes[1, 1].text(0.5, 0.5, 'Empty Plot', ha='center', va='center')
axes[1, 1].set_axis_off()

# Overall figure title
fig.suptitle('2x2 Subplot Example')

# Layout adjustments
plt.tight_layout()
plt.subplots_adjust(top=0.88, wspace=0.3, hspace=0.4)

# Save the figure
fig.savefig('subplot_figure.png')

# Show the plot
plt.show()
```

```python
# Example data
x = np.linspace(0, 10, 50)
y = np.sin(x)
np.random.seed(0)
data = np.random.randn(1000)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
scatter_x = np.random.rand(100)
scatter_y = np.random.rand(100)
matrix_data = np.random.rand(10, 12)
violin_data = [np.random.normal(loc, 1.0, 100) for loc in [1, 2, 3, 4]]
```

```python
# Create figure with subplots
fig, axs = plt.subplots(4, 3, figsize=(18, 16))
fig.suptitle('Various Types of Plots (Matplotlib Only)', fontsize=18)

# Line Plot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Plot')

# Scatter Plot
axs[0, 1].scatter(scatter_x, scatter_y, color='green')
axs[0, 1].set_title('Scatter Plot')

# Bar Chart
axs[0, 2].bar(categories, values, color='orange')
axs[0, 2].set_title('Bar Chart')

# Histogram
axs[1, 0].hist(data, bins=20, color='skyblue')
axs[1, 0].set_title('Histogram')

# Pie Chart
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Box Plot
axs[1, 2].boxplot(data)
axs[1, 2].set_title('Box Plot')

# Area Plot
axs[2, 0].fill_between(x, y, color='purple', alpha=0.5)
axs[2, 0].set_title('Area Plot')

# Stem Plot
axs[2, 1].stem(x[:10], y[:10], basefmt=" ")
axs[2, 1].set_title('Stem Plot')

# Heatmap
heatmap = axs[2, 2].imshow(matrix_data, cmap='YlGnBu', aspect='auto')
axs[2, 2].set_title('Heatmap')
fig.colorbar(heatmap, ax=axs[2, 2])

# Violin Plot
parts = axs[3, 0].violinplot(violin_data, showmeans=False, showmedians=True)
axs[3, 0].set_title('Violin Plot')

# Hexbin Plot
hb = axs[3, 1].hexbin(scatter_x, scatter_y, gridsize=25, cmap='inferno')
axs[3, 1].set_title('Hexbin Plot')
fig.colorbar(hb, ax=axs[3, 1])

# Error Bars
axs[3, 2].errorbar(x[:10], y[:10], yerr=0.2, fmt='-o')
axs[3, 2].set_title('Error Bars')

# Layout adjustment
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Styling and Customization

# Sample data
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Change background color
fig.patch.set_facecolor('#f2f2f2')

# Plot multiple lines with different styles
ax.plot(x, y1, linestyle='-', marker='o', color='red', alpha=0.8, label='Sine Wave')
ax.plot(x, y2, linestyle='--', marker='^', color='#007acc', alpha=0.7, label='Cosine Wave')

# Labels and title with custom font size
ax.set_title('Customized Line Plot', fontsize=16)
ax.set_xlabel('X Axis', fontsize=12)
ax.set_ylabel('Y Axis', fontsize=12)

# Set axis limits
ax.set_xlim([0, 10])
ax.set_ylim([-1.5, 1.5])

# Customize tick marks
ax.set_xticks(np.arange(0, 11, 2))
ax.set_yticks(np.arange(-1.5, 2.0, 0.5))

# Show grid with custom style
ax.grid(True, linestyle=':', linewidth=0.7, color='gray', alpha=0.7)

# Add legend with location
ax.legend(loc='upper right', fontsize=10)

# Adjust subplot spacing
plt.subplots_adjust(left=0.1, right=0.95, top=0.9, bottom=0.15)

# Show plot
plt.show()
```

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, label='Sine Wave', color='purple')

# Add labels and legend
ax.set_title('Sine Wave Plot')
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
ax.legend()

# Tight layout before saving
plt.tight_layout()

# Save to PNG with high resolution and tight bounding box
plt.savefig('sine_wave.png', dpi=300, bbox_inches='tight')

# Save with transparent background
plt.savefig('sine_wave_transparent.png', dpi=300, bbox_inches='tight', transparent=True)

# Save as PDF (vector format)
plt.savefig('sine_wave.pdf', bbox_inches='tight')

# Save a version without axes
ax.set_axis_off()
plt.savefig('sine_wave_no_axes.png', dpi=300, bbox_inches='tight')

# Display plot
plt.show()
```
